Dictonary Operations

```python
dict={"name":"harsh","age":23}
dict
```

```python
#get() ‚Äî dictionary ‡§Æ‡§ß‡•Ç‡§® key ‡§ö‡•Ä value ‡§∏‡•Å‡§∞‡§ï‡•ç‡§∑‡§ø‡§§‡§™‡§£‡•á access ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22}

# existing key
print("Name:", student.get("name"))

# non-existing key without default
print("City:", student.get("city"))

# non-existing key with default
print("City with default:", student.get("city", "Pune"))
```

```python
#items() ‚Äî dictionary ‡§Æ‡§ß‡•Ä‡§≤ ‡§∏‡§∞‡•ç‡§µ key-value pairs tuple ‡§ö‡•ç‡§Ø‡§æ form ‡§Æ‡§ß‡•ç‡§Ø‡•á access ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}
items = student.items()
print("Items:", items)
```

```python
#keys() ‚Äî dictionary ‡§Æ‡§ß‡•Ä‡§≤ ‡§∏‡§∞‡•ç‡§µ keys access ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}

keys = student.keys()
print("Keys:", keys)
```

```python
#popitem() ‚Äî dictionary ‡§Æ‡§ß‡•Ç‡§® last inserted key-value pair ‡§ï‡§æ‡§¢‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}
last_item = student.popitem()
last_item
```

```python
#clear() ‚Äî dictionary ‡§Æ‡§ß‡•Ä‡§≤ ‡§∏‡§∞‡•ç‡§µ items ‡§π‡§ü‡§µ‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}
student.clear()
student
```

```python
#copy() ‚Äî dictionary ‡§ö‡•Ä shallow copy ‡§¨‡§®‡§µ‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}
student_copy = student.copy()
student_copy
```

```python
#fromkeys() ‚Äî ‡§¶‡§ø‡§≤‡•á‡§≤‡•ç‡§Ø‡§æ keys ‡§ö‡•Ä dictionary ‡§§‡§Ø‡§æ‡§∞ ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§ ‡§Ü‡§£‡§ø ‡§∏‡§∞‡•ç‡§µ values ‡§è‡§ï‡§∏‡§æ‡§∞‡§ñ‡•ç‡§Ø‡§æ set ‡§ï‡§∞‡§§‡§æ‡§§.

keys = ["name", "age", "city"]
student = dict.fromkeys(keys)
student 
```

```python
#pop() ‚Äî dictionary ‡§Æ‡§ß‡•Ç‡§® specific key ‡§ï‡§æ‡§¢‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§Ü‡§£‡§ø ‡§§‡•ç‡§Ø‡§æ‡§ö‡•Ä value return ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.

student = {"name": "Harsh", "age": 22, "city": "Pune"}

age = student.pop("age")
print("Popped age:", age)
student
```

```python
#setdefault() ‚Äî dictionary ‡§Æ‡§ß‡•ç‡§Ø‡•á key ‡§ö‡§æ value access ‡§ï‡§ø‡§Ç‡§µ‡§æ set ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§, ‡§ú‡§∞ key ‡§®‡§∏‡•á‡§≤ ‡§§‡§∞ default value assign ‡§ï‡§∞‡§§‡•ã.

student = {"name": "Harsh", "age": 22}
name = student.setdefault("name", "Unknown")
print("Name:", name)
student
```

```python
#update() ‚Äî dictionary ‡§Æ‡§ß‡•ç‡§Ø‡•á ‡§è‡§ï ‡§ï‡§ø‡§Ç‡§µ‡§æ ‡§Ö‡§ß‡§ø‡§ï key-value pairs add/update ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.


student = {"name": "Harsh", "age": 22}

# update existing key
student.update({"age": 23})
print("After updating age:", student)

# add new key
student.update({"city": "Pune"})
print("After adding city:", student)

# update multiple keys at once
student.update({"marks": 90, "grade": "A"})
print("After adding marks and grade:", student)
```

```python
#values() ‚Äî dictionary ‡§Æ‡§ß‡•Ä‡§≤ ‡§∏‡§∞‡•ç‡§µ values access ‡§ï‡§∞‡§£‡•ç‡§Ø‡§æ‡§∏‡§æ‡§†‡•Ä ‡§µ‡§æ‡§™‡§∞‡§§‡§æ‡§§.


student = {"name": "Harsh", "age": 22, "city": "Pune"}

values = student.values()
print("Values:", values)
```

```python
# 1Ô∏è‚É£ Dictionary ‡§§‡§Ø‡§æ‡§∞ ‡§ï‡§∞‡§£‡•á
student = {"name": "Harsh", "age": 22, "city": "Pune"}
print("Original dictionary:", student)

# 2Ô∏è‚É£ Value access ‡§ï‡§∞‡§£‡•á
print(student["name"])               # Harsh
print(student.get("age"))            # 22
print(student.get("country", "N/A")) # key ‡§®‡§∏‡•á‡§≤ ‡§§‡§∞ default

# 3Ô∏è‚É£ Add / Update items
student["marks"] = 90                # ‡§®‡§µ‡•Ä‡§® key add
student["city"] = "Mumbai"           # value update
print("Add/Update ‡§®‡§Ç‡§§‡§∞:", student)

# 4Ô∏è‚É£ Items remove ‡§ï‡§∞‡§£‡•á
student.pop("age")                    # specific key remove
student.pop("country", "N/A")        # safe removal default ‡§∏‡§π
student.popitem()                     # last inserted item remove
print("Remove ‡§®‡§Ç‡§§‡§∞:", student)

# 5Ô∏è‚É£ Dictionary ‡§ö‡•Ä ‡§Æ‡§æ‡§π‡§ø‡§§‡•Ä ‡§Æ‡§ø‡§≥‡§µ‡§£‡•á
print(student.keys())                 # ‡§∏‡§∞‡•ç‡§µ keys
print(student.values())               # ‡§∏‡§∞‡•ç‡§µ values
print(student.items())                # ‡§∏‡§∞‡•ç‡§µ key-value pairs

# 6Ô∏è‚É£ Key existence check ‡§ï‡§∞‡§£‡•á
print("name" in student)              # True
print("age" in student)               # False

# 7Ô∏è‚É£ Dictionary clear ‡§ï‡§∞‡§£‡•á
# student.clear()
# print(student)

# 8Ô∏è‚É£ Dictionary copy ‡§ï‡§∞‡§£‡•á
student_copy = student.copy()
print("Copy:", student_copy)

# 9Ô∏è‚É£ fromkeys() -> keys ‡§™‡§æ‡§∏‡•Ç‡§® dictionary ‡§§‡§Ø‡§æ‡§∞ ‡§ï‡§∞‡§£‡•á
keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)    # ‡§∏‡§∞‡•ç‡§µ values default 0
print("Fromkeys:", new_dict)

# üîü setdefault() -> value access ‡§ï‡§ø‡§Ç‡§µ‡§æ default set ‡§ï‡§∞‡§£‡•á
val = student.setdefault("country", "India")
print("Setdefault country:", val)
print("Setdefault ‡§®‡§Ç‡§§‡§∞:", student)

# 1Ô∏è‚É£1Ô∏è‚É£ update() -> ‡§è‡§ï‡§æ‡§ö ‡§µ‡•á‡§≥‡•Ä multiple key-value add/update ‡§ï‡§∞‡§£‡•á
student.update({"age": 23, "grade": "A"})
print("Update ‡§®‡§Ç‡§§‡§∞:", student)
```
