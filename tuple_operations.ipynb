Tuple Operations

```python
tuple=(10,20,30,40,50)
tuple
```

```python
#count() — Tuple मध्ये दिलेला value किती वेळा आला आहे हे मोजण्यासाठी वापरतात.

tuple = (10, 20, 10, 30, 10)
print(tuple.count(10))
```

```python
#index() — Tuple मध्ये दिलेला value पहिल्यांदा कुठल्या index वर आहे ते शोधण्यासाठी वापरतात.

tuple = (10, 20, 30, 40)

print(tuple.index(30))
```

```python
# 1. Tuple तयार करणे
t = (10, 20, 30)          # () वापरून tuple तयार करतात
print("Tuple:", t, "\n")

# 2. Single element Tuple
t1 = (10,)                # 1 element असेल तर शेवटी comma आवश्यक
print("Single element tuple:", t1, "\n")

# 3. Indexing
t = (10, 20, 30)
print("Indexing examples:")
print(t[0])               # 0 index → पहिला element
print(t[2], "\n")         # 2 index → तिसरा element

# 4. Negative Indexing
print("Negative indexing:")
print(t[-1])              # -1 → शेवटचा element
print(t[-2], "\n")        # -2 → शेवटून दुसरा element

# 5. Slicing
t = (10, 20, 30, 40, 50)
print("Slicing t[1:4]:")
print(t[1:4], "\n")       # index 1 ते 3 elements print होतात

# 6. Length
print("Length of tuple:")
print(len(t), "\n")       # tuple मध्ये किती elements आहेत ते

# 7. count()
t = (10, 20, 10, 30, 10)
print("Count of 10:")
print(t.count(10), "\n")  # दिलेला value किती वेळा आहे ते मोजतो

# 8. index()
t = (10, 20, 30, 40)
print("Index of 30:")
print(t.index(30), "\n")  # दिलेला value पहिल्यांदा कुठे आहे ते

# 9. Concatenation
t1 = (1, 2)
t2 = (3, 4)
print("Concatenation:")
print(t1 + t2, "\n")      # दोन tuples एकत्र जोडतो

# 10. Repetition
t = (10, 20)
print("Repetition (t * 3):")
print(t * 3, "\n")        # tuple ची repeat copy करतो

# 11. Membership Test
t = (10, 20, 30)
print("Membership Test:")
print(20 in t)            # 20 tuple मध्ये आहे का?
print(50 not in t, "\n")  # 50 नाही का? True

# 12. min(), max(), sum()
t = (10, 5, 30)
print("min, max, sum:")
print(min(t))             # सर्वात लहान
print(max(t))             # सर्वात मोठा
print(sum(t), "\n")       # सर्व elements ची बेरीज

# 13. sorted()
t = (40, 10, 30, 20)
print("sorted() gives list:")
print(sorted(t))          # tuple sort करून list देतो
print(t, "\n")            # tuple बदलत नाही

# 14. Unpacking
t = (10, 20, 30)
a, b, c = t               # tuple मधील values variables मध्ये टाकतो
print("Unpacking:")
print(a, b, c, "\n")

# 15. Nested Tuple
t = (10, (20, 30), 40)
print("Nested tuple access:")
print(t[1][0], "\n")      # inner tuple मधून value काढतो

# 16. Tuple immutable example
print("Immutable Example:")
try:
    t = (1, 2, 3)
    t[0] = 10             # Tuple मध्ये बदल allowed नाही
except Exception as e:
    print(e, "\n")

# 17. Tuple → List → Modify → Tuple
t = (10, 20, 30)
lst = list(t)             # tuple → list convert (changing allowed)
lst.append(40)            # list मध्ये add करता येते
t = tuple(lst)            # परत tuple बनवतो
print("After convert and modify:", t)
```
