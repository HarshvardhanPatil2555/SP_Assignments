```python
#isupper() → string मधील सर्व letters uppercase आहेत का ते तपासतो.

str1.isupper()
```

```python
#join() method चा उपयोग iterable मधील elements (जसे की list, tuple) एकत्र करून एक single string तयार करण्यासाठी केला जातो, आणि str ही separator म्हणून वापरली जाते.

str9 = "- harshu -"  # separator
my_list = ["Hello", "World"]

result = str9.join(my_list)

print(result)
```

```python
#ljust() मेथडचा वापर string ला डावीकडे (left) justify करण्यासाठी केला जातो.

str10 = "Hello"

# डावीकडे justify, right मध्ये space भरली जाते
print(str10.ljust(10))        # Output: 'Hello     '

# custom character वापरून
print(str10.ljust(10, '*'))   # Output: 'Hello*****'
```

```python
#lower() मेथडचा वापर string मधील सर्व अक्षरे लहान (lowercase) मध्ये बदलण्यासाठी केला जातो.

str1.lower()
```

```python
#lstrip() मेथडचा वापर string च्या डाव्या (left) बाजूतील रिकाम्या जागा (spaces) किंवा specified characters काढण्यासाठी केला जातो.

str11 = "xxxHello Worldxxx"
print(str11.lstrip('x'))
```

```python
#maketrans() ही method translation table तयार करण्यासाठी वापरली जाते, जी नंतर str.translate() सोबत वापरली जाते. ही method मूळ string बदलत नाही; फक्त mapping तयार करते.

# Translation table तयार करणे
table = str.maketrans('abc', '123')

# translate वापरून बदल करणे
text = "abcde"
print(text.translate(table))

#'a'→'1', 'b'→'2', 'c'→'3' बदलले, इतर characters जसाचे तसाचे राहिले.
```

```python
#partition() मेथडचा वापर string ला पहिल्या दिलेल्या separator वर split करून तीन भागात विभागण्यासाठी केला जातो.

text = "hello world python"
result = text.partition("hello")
print(result)
```

```python
#removeprefix() मेथडचा वापर string च्या सुरुवातीला (prefix) असलेला specified substring काढण्यासाठी केला जातो.

text = "HelloWorld"
print(text.removeprefix("Hello"))
```

```python
#replace() मेथडचा वापर string मध्ये एक substring दुसऱ्या substring ने बदलण्यासाठी केला जातो.

text = "hello world, world is big"
new_text = text.replace("world", "Python")
print(new_text)
```

```python
#removesuffix() मेथडचा वापर string च्या शेवटी (suffix) असलेला specified substring काढण्यासाठी केला जातो.

text = "HelloWorld"
print(text.removesuffix("World"))
```

```python
#rfind() मेथडचा वापर string मध्ये दिलेले substring शेवटून (right side) शोधण्यासाठी केला जातो आणि त्याचा index परत करतो.

text = "hello world, hello python"
index = text.rfind("hello")
print(index)

#TIP:-rfind() आणि find() मधला फरक म्हणजे find() string च्या सुरुवातीपासून शोधतो, तर rfind() शेवटून शोधतो
```

```python
#rindex() मेथड string मध्ये दिलेल्या substring ला शेवटून शोधते आणि त्याचा index परत करते.

text = "hello world, hello python"
index = text.rindex("hello")
print(index)
```

```python
#rsplit() मेथडचा वापर string ला right side पासून split करून list मध्ये रूपांतर करण्यासाठी केला जातो.

text = "one two three four"
print(text.rsplit())
```

```python
#rstrip() मेथडचा वापर string च्या उजव्या (right) बाजूतील रिकाम्या जागा (spaces) किंवा specified characters काढण्यासाठी केला जातो.

str1 = "   Hello World   "
print(str1.rstrip())  #'   Hello World'
```

```python
#split() मेथडचा वापर string ला list मध्ये भाग करण्यासाठी (split करण्यासाठी) केला जातो.

text = "one two three four"
print(text.split())
```

```python
#splitlines() ही Python मधील string method आहे. तिचा उपयोग string ला lines (ओळी) मध्ये विभाजित करण्यासाठी होतो, म्हणजेच प्रत्येक newline (\n, \r, किंवा \r\n) वर string वेगवेगळी element म्हणून list मध्ये बदलते.

text = "Hello\nWorld\nPython"
lines = text.splitlines()
print(lines)
```

```python
#startswith() ही Python मधील string method आहे जी तपासते की string कोणत्या specific prefix (अर्थात सुरुवातीच्या अक्षर/अक्षरे) ने सुरू होते की नाही.

str12= "Python Programming"

# single prefix
print(str12.startswith("Python"))   # True

# wrong prefix
print(str12.startswith("Java"))     # False

# with start and end
print(str12.startswith("Programming", 7))  # True, index 7 पासून "Programming" सुरू होते
```

```python
#strip() ही Python मधील string method आहे जी string च्या दोन्ही बाजूंनी असलेली whitespace किंवा specific characters काढून टाकते.

str2 = "###Python###"
cleaned = str2.strip("#")
print(cleaned)
```

```python
#swapcase() ही Python मधील string method आहे जी string मधील प्रत्येक अक्षराचे case उलट करते. म्हणजे, uppercase → lowercase आणि lowercase → uppercase होते.

str1 = "Hello World 123"
swapped = str1.swapcase()
print(swapped)
```

```python
#title() ही Python मधील string method आहे जी string मधील प्रत्येक शब्दाचा survātīla अक्षर (first letter) मोठा (uppercase) करतो आणि उर्वरित अक्षरे लहान (lowercase) करतो.

str1 = "hello world from python"
title_str = str1.title()
print(title_str)
```

```python
#translate() ही Python मधील string method आहे जी string मधील characters replace किंवा delete करण्यासाठी वापरली जाते, पण ही method स्वतः standalone वापरली जात नाही; ती translation table (mapping) सोबत वापरली जाते.

# Translation table तयार करणे
table = str.maketrans("aeiou", "12345")  # vowels → numbers

str1 = "harshvardhan"
translated = str1.translate(table)
print(translated)
```

```python
#upper() ही Python मधील string method आहे जी string मधील सर्व अक्षरे मोठ्या (uppercase) अक्षरांमध्ये बदलते.

str1 = "hello world 123"
upper_str = str1.upper()
print(upper_str)
```

```python
#zfill() ही Python मधील string method आहे जी string च्या left side मध्ये zeros (0) भरून string ची desired length पूर्ण करते. हे numeric strings किंवा ID formatting साठी उपयोगी आहे.

num = "42"
padded = num.zfill(8)
print(padded)
```

```python
#__add__() ही Python मधील special (dunder) method आहे जी string concatenation साठी वापरली जाते.सामान्यतः आपण + operator वापरतो, पण अंतर्गत Python मध्ये + हे __add__() method कॉल करून काम करते.

str1 = "Hello "
str2 = "World"

# Using + operator
print(str1 + str2)  

# Using __add__() method
print(str1.__add__(str2))
```

```python
#__class__ ही Python मधील special (dunder) attribute आहे जी सांगते की एखाद्या object चा type/class कोणता आहे.

str1 = "Hello World"
print(str1.__class__)
```
