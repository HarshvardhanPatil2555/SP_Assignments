List Operations:-

```python
lst = [1,2,3,4,5,6,7,8]
lst
```

```python
#index() हा List मधील एखाद्या value चा index (स्थान क्रमांक) शोधण्यासाठी वापरला जाणारा method आहे.

lst = [10, 20, 30, 40, 50]

# 30 चा index शोधतो
result = lst.index(30)
result
```

```python
#reverse() → लिस्ट उलटी करते, मूळ लिस्ट बदलते, काहीही return करत नाही.

lst = [1, 2, 3, 4, 5]

lst.reverse()
print("After :", lst)
```

```python
#append(x) → लिस्टच्या शेवटी नवीन element x add करते. मूळ लिस्ट बदलते.

lst = [10, 20, 30]

lst.append(40)   # add 40
print
```

```python
#clear() → लिस्टमधले सगळे elements काढून लिस्ट रिकामी करते.

lst = [10, 20, 30, 40]

lst.clear()
lst
```

```python
#copy() → लिस्टची एक नवीन शॅलो कॉपी तयार करते.

lst = [10, 20, 30]
new_lst = lst.copy()

new_lst
```

```python
#count()->लिस्टमध्ये दिलेला element किती वेळा येतो ते मोजून number परत करते.

lst = [10, 20, 10, 30, 10, 40]

lst.count(10)
```

```python
#extend()->एका लिस्टमधील सगळे elements दुसऱ्या लिस्टच्या शेवटी add करते. (मूळ लिस्ट बदलते)

lst = [1, 2, 3]
lst.extend([4, 5, 6])
lst
```

```python
#extend() → दुसऱ्या लिस्टमधले सगळे elements मूळ लिस्टच्या शेवटी add करते.

lst = [10, 20, 30]
lst.extend([40, 50])
lst
```

```python
#index() ->ठराविक index वर element add करते आणि लिस्ट shift होते.
lst = [10, 20, 30]
lst.insert(1, 99)   # index 1 वर 99 add

lst
```

```python
#pop() ->दिलेल्या indexवरील element काढून टाकते आणि तो element return करते. (index दिला नाही तर शेवटचा element काढते)

lst = [10, 20, 30, 40]

x = lst.pop()     # शेवटचा element काढतो
print("Popped:", x)
print("List:", lst)
lst = [10, 20, 30, 40]

y = lst.pop(1)    # index 1 वरील element (20) काढतो
print("Popped:", y)
print("List:", lst)
```

```python
#remove() — list मधून दिलेला value काढण्यासाठी वापरला जातो.

lst = [10, 20, 30, 20, 40]

lst.remove(20)     # पहिला 20 काढतो
print(lst)
```

```python
#sort() — list मधील elements ascending (लहान → मोठे) क्रमात लावण्यासाठी वापरला जातो.

lst = [40, 10, 30, 20]
lst.sort()
print(lst) # Accending order

lst = [40, 10, 30, 20]
lst.sort(reverse=True)
print(lst) # Descending order
```
