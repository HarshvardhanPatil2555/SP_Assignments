```python
set={10,20,30,40,50,60}
set
```

```python
#add()->set मध्ये एक नवीन element टाकण्यासाठी वापरतात.

set = {10, 20, 30}
print("Original Set :", set)

# add() वापरून element टाकणे
set.add(40)
print("After add(40) :", set)

# duplicate element add करणे (काही फरक नाही)
set.add(20)
print("After add(20) again :", set)
```

```python
#clear() set मधील सर्व elements काढून टाकून set पूर्णपणे रिकामा करतो.

set = {10, 20, 30}
print("Original Set :", set)

set.clear()
print("After clear() :", set)
```

```python
#copy() — set ची कॉपी (duplicate) बनवण्यासाठी वापरतात.


my_set = {10, 20, 30}

new_set = my_set.copy()   # कॉपी तयार

print("Original Set :", my_set)
print("Copied Set   :", new_set)
```

```python
#discard() — set मधून दिलेला element काढून टाकतो, पण element नसला तरी error देत नाही.

my_set = {10, 20, 30, 40}

my_set.discard(20)     # 20 काढेल
print("After discard(20):", my_set)

my_set.discard(100)    # 100 नसल्यामुळे काही होत नाही
print("After discard(100):", my_set)
```

```python
#intersection() — दोन किंवा अधिक sets मधील common elements (intersection) शोधण्यासाठी वापरतात.

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

common = set1.intersection(set2)

common
```

```python
#intersection_update() — दोन किंवा अधिक sets मधील common elements ठेवण्यासाठी वापरतात आणि मूळ set in-place update करतो.

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

set1.intersection_update(set2)  # set1 मध्ये फक्त common elements राहतील

print("Set1 after intersection_update:", set1)
print("Set2 remains unchanged:", set2)
```

```python
#issubset() — एक set दुसऱ्या set मध्ये पूर्णपणे समाविष्ट आहे का ते तपासण्यासाठी वापरतात.

set1 = {10, 20}
set2 = {10, 20, 30, 40}

result = set1.issubset(set2)
print("Is set1 a subset of set2?:", result)

# Example where False
set3 = {10, 50}
print("Is set3 a subset of set2?:", set3.issubset(set2))
```

```python
#pop() set मधून कोणताही एक element काढतो (random, कारण set unordered आहे).


my_set = {10, 20, 30, 40}
my_set.pop()
```

```python
#remove() — set मधून specific element काढण्यासाठी वापरतात.


my_set = {10, 20, 30, 40}
my_set.remove(20)     # 20 काढतो
my_set
```

```python
#symmetric_difference() — दोन sets मधील unique elements (ज्या फक्त एका set मध्ये आहेत) परत करण्यासाठी वापरतात.

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

unique_elements = set1.symmetric_difference(set2)
unique_elements
```

```python
#symmetric_difference_update() — दोन sets मधील unique elements (ज्या फक्त एका set मध्ये आहेत) ठेवण्यासाठी वापरतात आणि मूळ set in-place update करतो.

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

set1.symmetric_difference_update(set2)

set1
```

```python
#union() — दोन किंवा अधिक sets मधील सर्व unique elements मिळवण्यासाठी वापरतात.

set1 = {10, 20, 30}
set2 = {30, 40, 50}

union_set = set1.union(set2)

union_set
```

```python
#update() — एक set दुसऱ्या set किंवा iterable मधील elements ने in-place update करण्यासाठी वापरतात.

set1 = {10, 20, 30}
set2 = {30, 40, 50}

set1.update(set2)  # set1 मध्ये set2 चे elements add

set1
```

```python
#isdisjoint() — दोन sets मध्ये कोणतेही common elements नाहीत का ते तपासण्यासाठी वापरतात.

set1 = {10, 20, 30}
set2 = {40, 50, 60}
set3 = {20, 70}

print("set1 and set2 disjoint?:", set1.isdisjoint(set2))
print("set1 and set3 disjoint?:", set1.isdisjoint(set3))
```

```python
#issuperset() — एक set दुसऱ्या set चा superset आहे का ते तपासण्यासाठी वापरतात.

set1 = {10, 20, 30, 40}
set2 = {20, 30}
set3 = {50, 60}

print("Is set1 a superset of set2?:", set1.issuperset(set2))
print("Is set1 a superset of set3?:", set1.issuperset(set3))
```

```python
#difference() — दोन sets मधील अंतर (elements जे फक्त पहिल्या set मध्ये आहेत, दुसऱ्या set मध्ये नाहीत) शोधण्यासाठी वापरतात.

set1 = {10, 20, 30, 40}
set2 = {30, 40, 50, 60}

diff = set1.difference(set2)
print("Set1:", set1)
print("Set2:", set2)
print("Difference (set1 - set2):", diff)
```
